```python
df_cleaned['Rating'] = pd.to_numeric(df_cleaned['Rating'], errors='coerce')
# Step 2: Convert 'Fee Start Range' and 'Fee End Range' to float
# Remove characters like 'K' and 'L' and convert to numeric values
df_cleaned['Fee Start Range'] = df_cleaned['Fee Start Range'].replace({' K': 'e3', ' L': 'e5'}, regex=True)
df_cleaned['Fee Start Range'] = pd.to_numeric(df_cleaned['Fee Start Range'], errors='coerce')

df_cleaned['Fee End Range'] = df_cleaned['Fee End Range'].replace({' K': 'e3', ' L': 'e5'}, regex=True)
df_cleaned['Fee End Range'] = pd.to_numeric(df_cleaned['Fee End Range'], errors='coerce')
```
